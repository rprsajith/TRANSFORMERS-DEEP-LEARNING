```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
classifier("sajith is a sports player")
```

```python
classifier = pipeline("sentiment-analysis")
classifier(["sajith is a sports player.", "but sajith is injured while playing."])
```

```python
classifier = pipeline("zero-shot-classification")
classifier("sajith is a sports player",
candidate_labels=["education", "sports", "fitness"])
```

```python
generator = pipeline("text-generation")
generator("sajith is a sports player of volleyball")
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator( "In my future",
          max_length = 30,
           num_return_sequences=2,
)
```

```python
unmasker = pipeline("fill-mask")
unmasker("sajith is sports <mask> of volleyball.", top_k=2)
```

```python
ner = pipeline('ner',grouped_entities=True)
ner("I am Sajith and I work at Kgisl in Coimbatore,In the position of Volleyball Coach")
```

```python
question_answer = pipeline("question-answering")
question_answer(
    question = "where I am going to play?",
    context="you go and play kgisl ground",
)
```
